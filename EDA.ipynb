```python
percentagedata=[]                                        #creating a empty list datatype to store data
for i in Avviation.columns:                               
    null_count=Avviation[i].isnull().sum()
    percentage=round((null_count/Avviation.shape[0])*100,2)
    percentagedata.append([i, null_count, percentage])
'''
using for loop and itrating to each row to get there null values dividing the null values by total number of rows
to get there percentage multipling the values with 100 for percentage
'''
Percentagadf=pd.DataFrame(percentagedata)
```

```python
Percentagadf.rename(columns={0:'Column Name',1:'Number of Null Values',2:'Percentage'},inplace=True) 
#renaming the columns to a relavent one using inplace=True to commit the changes
Percentagadf
```

```python
filt=Percentagadf['Percentage']>=30
Morethan30=Percentagadf[filt]
Morethan30
# using filt and storing the filter where null percentage of column is more than 30%
```

```python
mean_mode_values = {
    'Total.Uninjured': Avviation['Total.Uninjured'].mean(),
    'Total.Minor.Injuries': Avviation['Total.Minor.Injuries'].mean(),
    'Total.Serious.Injuries': Avviation['Total.Serious.Injuries'].mean(),
    'Total.Fatal.Injuries': Avviation['Total.Fatal.Injuries'].mean(),
    'Purpose.of.flight':Avviation['Purpose.of.flight'].mode()[0],
    'Air.carrier':Avviation['Air.carrier'].mode()[0],
    'Aircraft.Category':Avviation['Aircraft.Category'].mode()[0],
    'FAR.Description':Avviation['FAR.Description'].mode()[0],
    'Schedule':Avviation['Schedule'].mode()[0],
    'Air.carrier':Avviation['Air.carrier'].mode()[0],
    'Broad.phase.of.flight':Avviation['Broad.phase.of.flight'].mode()[0],
    'Engine.Type':Avviation['Engine.Type'].mode()[0],
    'Number.of.Engines':int(Avviation['Number.of.Engines'].mean()),
    'Weather.Condition':Avviation['Weather.Condition'].mode()[0],
    'Airport.Code':Avviation['Airport.Code'].mode()[0],
    'Airport.Name':Avviation['Airport.Name'].mode()[0],
    'Aircraft.damage':Avviation['Aircraft.damage'].mode()[0],
    'Injury.Severity':Avviation['Injury.Severity'].mode()[0],
    'Make':Avviation['Make'].mode()[0],
    'Amateur.Built':Avviation['Amateur.Built'].mode()[0],
    'Model':Avviation['Model'].mode()[0],
    'Report.Status':Avviation['Report.Status'].mode()[0],
    'Registration.Number':Avviation['Registration.Number'].mode()[0],
    'Country':Avviation['Country'].mode()[0],
    'Location':Avviation['Location'].mode()[0]
}
Avviation.fillna(value=mean_mode_values, inplace=True)
# cleaning data and replaing the data with the relavent method for object's and numerics 
# replacing mean for numerical null values and mode for object types
```

```python
zscore_col=Avviation.select_dtypes('number')
Avviationzscore=Avviation.copy()
for i in zscore_col:
    Avviationzscore[f'{i}_Z'] = zscore(Avviationzscore[i])
outliers=Avviationzscore[(Avviationzscore[[f'{i}_Z' for i in zscore_col]] > 3).any(axis=1) | (Avviationzscore[[f'{i}_Z' for i in zscore_col]] < -3).any(axis=1)]
print(outliers)
# calculating zscore value for numerical data in the dataset 
```

```python
plt.figure(figsize=(15, 10))
for i,column in enumerate(zscore_col, start=1):
    plt.subplot(5,2, i)  
    sns.boxplot(x=Avviation[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
plt.tight_layout()
plt.show()
# using box plot to identify the outliers in the dataset for zscore values
```

```python
Weather_Accident=Avviation.groupby('Weather.Condition')['Accident.Number'].size().rename('Accident_count').reset_index()
# grouping the weather condition and counting the most number of accidents that happened during which weather condition
```

By this analysis we can conclude that most accident happend when wheather is VMC which stands for -VMC (Visual Meteorological Conditions) refer to weather conditions that allow a pilot to fly visually, meaning they can rely on their eyesight to maintain safe separation from terrain and other aircraft.

```python
Avviation['Fatality.Rate']=(Avviation['Total.Fatal.Injuries']/
                                  (Avviation['Total.Fatal.Injuries'] +
                                   Avviation['Total.Serious.Injuries'] +
                                   Avviation['Total.Minor.Injuries'] +
                                   Avviation['Total.Uninjured'])) * 100
Avviation['Fatality.Rate']=Avviation['Fatality.Rate'].fillna(0)
```

```python
FacFa=Avviation.groupby('Number.of.Engines')['Fatality.Rate'].mean().rename('Fatality.Rate').reset_index()
```

```python
corr_matrix=FacFa[['Number.of.Engines','Fatality.Rate']].corr()
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.title('Relationship Between Engine Number and Fatality rate')
plt.show()
```

The correlation between Fatality rate and Engine Count is is negtively impaced
